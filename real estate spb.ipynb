–¶–∏–∫–ª–æ–º –∑–∞–ø–æ–ª–Ω—è–µ–º –≤—Å–µ –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è –≤ —Å—Ç–æ–ª–±—Ü–µ –ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏ –Ω–∞ –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ —Å—Ä–µ–¥–Ω–µ–≥–æ —Å–æ–æ—Ç–Ω–æ—à–µ–Ω–∏—è –∏ –æ–±—â–µ–π –ø–ª–æ—â–∞–¥–∏ –∫–≤–∞—Ä—Ç–∏—Ä—ã

```python
# —Ü–∏–∫–ª, –∑–∞–ø–æ–ª–Ω—è—é—â–∏–π –ø—Ä–æ–ø—É—Å–∫–∏ 
for value in data['living_area'].isna():
    data.loc[data['living_area'].isna(), 'living_area'] = data['total_area'] * 0.56
```

```python
# –º–µ—Ç–æ–¥–æ–º corr –Ω–∞—Ö–æ–¥–∏–º –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—é –ü–∏—Ä—Å–æ–Ω–∞
data['kitchen_area'].corr(data['total_area'])
```

–ó–¥–µ—Å—å –∫–æ—Ä—Ä–µ–ª—è—Ü–∏—è –Ω–µ —Å–∏–ª—å–Ω–æ –≤—ã—Ä–∞–∂–µ–Ω–∞. –ó–Ω–∞—á–∏—Ç –±—É–¥–µ–º –∑–∞–ø–æ–ª–Ω—è—Ç—å —Å –ø–æ–º–æ—â—å—é –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏: –æ–±—â–∞—è –ø–ª–æ—â–∞–¥—å * 0.9 - (–ø–ª–æ—â–∞–¥—å –∫—É—Ö–Ω–∏ + –ø–ª–æ—â–∞–¥—å –æ–±—â–∞—è)

–ó–¥–µ—Å—å —É–º–Ω–æ–∂–∞–µ–º –Ω–∞ 0.9, –ø–æ—Ç–æ–º—É —á—Ç–æ 10% –ø–ª–æ—â–∞–¥–∏ –æ—Å—Ç–∞–≤–ª—è–µ–º –¥–ª—è –Ω–µ–∂–∏–ª—ã—Ö –ø–æ–º–µ—â–µ–Ω–∏–π –∏ —Å—Ç–µ–Ω

–î–ª—è –Ω–∞—á–∞–ª–∞ –ø—Ä–æ–≤–µ—Ä–∏–º —Å–∫–æ–ª—å–∫–æ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ —Å—Ç—Ä–æ–∫–∞—Ö living_area –∏ total_area —Ç–∞–º, –≥–¥–µ —É–∂–µ –µ—Å—Ç—å –ø—Ä–æ–ø—É—Å–∫–∏ –≤ kitchen_area

```python
# –¥–µ–ª–∞–µ–º —Å—Ä–µ–∑ –¥–∞–Ω–Ω—ã—Ö –∏ –≤—ã–≤–æ–¥–∏–º —Å—É–º–º—É –ø—Ä–æ–ø—É—Å–∫–æ–≤ –≤ –∏–Ω—Ç–µ—Ä–µ—Å—É—é—â–∏—Ö –Ω–∞—Å –¥–∞–Ω–Ω—ã—Ö
data.query('kitchen_area.isna()')[['living_area','total_area']].isna().sum()
```

## –†–∞—Å—à–∏—Ä–∏–º —Ç–∞–±–ª–∏—Ü—É 

–î–æ–±–∞–≤–∏–º –≤ —Ç–∞–±–ª–∏—Ü—É –Ω–æ–≤—ã–µ —Å—Ç–æ–ª–±—Ü—ã —Å–æ —Å–ª–µ–¥—É—é—â–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏

–°—Ç–æ–ª–±–µ—Ü —Å —Ü–µ–Ω–æ–π –æ–¥–Ω–æ–≥–æ –∫–≤–∞–¥—Ä–∞—Ç–Ω–æ–≥–æ –º–µ—Ç—Ä–∞ —Å –æ–∫—Ä—É–≥–ª–µ–Ω–∏–µ–º –¥–æ –¥–≤—É—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π

```python
# –ø–æ–¥–µ–ª–∏–º —Å—Ç–æ–∏–º–æ—Å—Ç—å –æ–±—ä–µ–∫—Ç–∞ –Ω–∞ –µ–≥–æ –æ–±—â—É—é –ø–ª–æ—â–∞–¥—å –∏ –∑–∞—Ç–µ–º –æ–∫—Ä—É–≥–ª–∏–º –¥–æ –¥–≤—É—Ö –∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ –∑–∞–ø—è—Ç–æ–π
data['sqm_price'] = (data['last_price'] / data['total_area']).round(2)
```

–°—Ç–æ–±–ª–µ—Ü —Å –¥–Ω–µ–º –Ω–µ–¥–µ–ª–∏ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è (0 ‚Äî –ø–æ–Ω–µ–¥–µ–ª—å–Ω–∏–∫, 1 ‚Äî –≤—Ç–æ—Ä–Ω–∏–∫ –∏ —Ç–∞–∫ –¥–∞–ª–µ–µ)

```python
# –¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –¥–Ω–µ–º –Ω–µ–¥–µ–ª–∏ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
data['weekday_published'] = data['first_day_exposition'].dt.weekday
```

–°—Ç–æ–ª–±–µ—Ü —Å –Ω–æ–º–µ—Ä–æ–º –º–µ—Å—è—Ü–∞ –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è

```python
# –¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –º–µ—Å—è—Ü–µ–º –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
data['month_published'] = data['first_day_exposition'].dt.month
```

–°—Ç–æ–ª–±–µ—Ü —Å –≥–æ–¥–æ–º –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è

```python
# –¥–æ–±–∞–≤–∏–º —Å—Ç–æ–ª–±–µ—Ü —Å –≥–æ–¥–æ–º –ø—É–±–ª–∏–∫–∞—Ü–∏–∏ –æ–±—ä—è–≤–ª–µ–Ω–∏—è
data['year_published'] = data['first_day_exposition'].dt.year
```

–°—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏–µ–π —ç—Ç–∞–∂–µ–π

–ù–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é, –∫–æ—Ç–æ—Ä–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∏—Ä—É–µ—Ç —ç—Ç–∞–∂–∏

```python
# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è —Å—Ç—Ä–æ–∫–∏, –≤–æ–∑–≤—Ä–∞—â–∞—é—â–∞—è –∫–∞—Ç–µ–≥–æ—Ä–∏—é —ç—Ç–∞–∂–∞
def categorize_floor(row):
    
    floor = row['floor']
    total_floors = row['floors_total']
    # –ø—Ä–æ–≤–µ—Ä–∏–º —á—Ç–æ —É–∫–∞–∑–∞–Ω–Ω—ã–π —ç—Ç–∞–∂ –Ω–µ –±–æ–ª—å—à–µ –ø–æ—Å–ª–µ–¥–Ω–µ–≥–æ —ç—Ç–∞–∂–∞ –≤ –¥–æ–º–µ
    if 0 < floor <= total_floors:
    
        if floor == 1:
            return '–ø–µ—Ä–≤—ã–π'
        if floor == total_floors:
            return '–ø–æ—Å–ª–µ–¥–Ω–∏–π'
        else:
            return '–¥—Ä—É–≥–æ–π'
    else:
        return '–Ω–µ–≤–µ—Ä–Ω–æ —É–∫–∞–∑–∞–Ω —ç—Ç–∞–∂'
```

```python
# –ø—Ä–∏–º–µ–Ω–∏–º —Ñ—É–Ω–∫—Ü–∏—é –∫ –Ω–æ–≤–æ–º—É —Å—Ç–æ–ª–±—Ü—É —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏ —ç—Ç–∞–∂–µ–π
data['floor_category'] = data.apply(categorize_floor, axis=1)
```

–°—Ç–æ–ª–±–µ—Ü —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≥–æ—Ä–æ–¥–∞ –≤ –∫–∏–ª–æ–º–µ—Ç—Ä–∞—Ö

```python
# —Å–æ–∑–¥–∞–¥–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å —Ä–∞—Å—Å—Ç–æ—è–Ω–∏–µ–º –¥–æ —Ü–µ–Ω—Ç—Ä–∞ –≤ –∫–º
# –ø–µ—Ä–µ–≤–æ–¥–∏–º –º–µ—Ç—Ä—ã –≤ –∫–º –∏ –æ–∫—Ä—É–≥–ª—è–µ–º –¥–æ —Ü–µ–ª—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π
data['cityCenter_distance_km'] = (data['cityCenters_nearest'] / 1000).round(0)
```
